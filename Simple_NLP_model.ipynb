```python
from datasets import load_dataset
import pandas as pd
from google.colab import drive
```

```python
drive.mount("/content/drive")
```

```python
raw_dataset = load_dataset("imdb")
```

```python
train, test = pd.DataFrame(raw_dataset["train"]), pd.DataFrame(raw_dataset["test"])
```

```python
import re
```

```python
def clean(s: str) -> str:
    s = re.sub(r"<.*?>", "", s)
    s = re.sub(r"[^a-zA-Z0-9\s]", " ", s)
    s = re.sub(r"\s+"," ", s).strip()
    return s.lower()
```

```python
for ds in (train, test):
    ds["text"] = ds["text"].astype(str).apply(clean)
    ds["label"] = ds["label"].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, val = train_test_split(train, test_size=0.1, random_state=42, stratify=train["label"])
```

```python
print(f"Train size: {len(train)}\nValidation size: {len(val)}\nTest size: {len(test)}")
```

```python
SAVE_DIR = "/content/drive/MyDrive/IMDB_cleaned_dataset/"
train.to_csv(SAVE_DIR+"cleaned_train.csv")
val.to_csv(SAVE_DIR+"cleaned_val.csv")
test.to_csv(SAVE_DIR+"cleaned_test.csv")
```
